
# ACTIVIDAD 3.
Con estos datos que ya tienes agrega una nueva columna para que puedas predecir, en esa columna ahora vas a tener el pa√≠s del programador. Ojo, recuerda que la regresi√≥n lineal simple solo trabaja con datos continuos, as√≠ que cada pa√≠s deber√°s represarlo con un n√∫mero, por ejemplo Ecuador = 1, Espa√±a = 2.


```python
pd.options.mode.chained_assignment = None 
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
dataset = pd.read_csv('salarios_1.csv')
dataset.head (5)
```

```python
dataset['pais']=np.random.randint(1, 7,dataset.shape[0])
for i in dataset.index:
    if dataset['Salario'][i] >= 15000 and dataset['Salario'][i] < 20000:
        dataset['pais'][i] = 1 
    elif dataset['Salario'][i] >= 20000 and dataset['Salario'][i] < 25000:
        dataset['pais'][i] = 2
    elif dataset['Salario'][i] >= 25000 and dataset['Salario'][i] < 30000:
        dataset['pais'][i] = 3
    elif dataset['Salario'][i] >= 30000 and dataset['Salario'][i] < 35000:
        dataset['pais'][i] = 4
    else:
        dataset['pais'][i] = 5

dataset
#print(dataset.head(5))
```

```python
x=dataset.iloc[:, 2:3].values
y=dataset.iloc[:, 1].values
```

```python
X_train,X_test,Y_train,Y_test= train_test_split(x,y,test_size=0.2,random_state=0)
X_train
```

```python
regresion= LinearRegression()
regresion.fit(X_train, Y_train)
vision_train= plt
vision_train.scatter(X_train, Y_train, color ='green')
vision_train.plot(X_train, regresion.predict(X_train), color='black')
vision_train.title('Salario VS pa√≠s (Entrenamiento)')
vision_train.xlabel('Pa√≠s')
vision_train.ylabel('Salario')
vision_train.show
```

```python
vision_train.scatter(X_test, Y_test, color ='red')
vision_train.plot(X_train, regresion.predict(X_train), color='black')
vision_train.title('Salario VS pa√≠s')
vision_train.xlabel('Pa√≠s')
vision_train.ylabel('Salario')
vision_train.show
```

```python
regresion.score(X_test, Y_test)
```
